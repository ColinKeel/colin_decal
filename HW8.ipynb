```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)

amplitudes = [0.5, 1, 1.5, 2, 2.5]
frequencies = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 6))

for A, w, in zip(amplitudes, frequencies):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, label=f"A={A}, w={w}")

plt.title("Multiple Sine Waves")
plt.xlabel("x")
plt.ylabel("y = A*sin(w*x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", ax=axes[0])
axes[0].set_title("Penguin Bill Dimensions")
axes[0].set_xlabel("Bill Length (mm)")
axes[0].set_ylabel("Bill Depth (mm)")
axes[0].legend(title="Species")

sns.histplot(data=penguins, x="body_mass_g", kde=False, ax=axes[1], bins=20, color='skyblue')
axes[1].set_title("Distribution of Penguin Body Mass")
axes[1].set_xlabel("Body Mass (g)")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
